```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import csv

sales = pd.read_csv("top100videogamesales.csv")
```

# Global Sales

The line chart below shows the total global sales from 2002 to 2014.

```python
sales.groupby('Year').sum().plot(kind='line', y='Global_Sales', color='green', ylabel='Total Global Sales')
```

The code below shows a table with the total global sales per year.

```python
print(sales.groupby('Year').sum().loc[:, "Global_Sales"])
```

From 2002 to 2014, 2006 experienced the highest amount of total global sales (167.44 USD), whereas 2003 experienced the lowest amount of total global sales (14.15 USD).

Below shows a pie chart showing the distribution of global sales of video games per platform.

```python
platform_colors = ['#1a00ff', '#c137c8', '#d5c02a', '#3fc077']
pie_chart = sales.groupby('Platform').sum().plot(kind='pie', y='Global_Sales', title='Distribution of Global Sales of Video Games by Platform', colors=platform_colors, figsize=(7,7), autopct='%1.1f%%')
pie_chart.yaxis.set_visible(False)
```

The code below shows a table of the total global sales per platform.

```python
print(sales.groupby('Platform').sum().loc[:, "Global_Sales"])
```

PlayStation games make up 27% of all global sales.

Below shows a stacked bar chart showing the sales by region for each publisher.

```python
total_global_sales = sales.groupby('Publisher').sum().loc[:, ["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]]
total_global_sales.plot.bar(stacked=True, title='Total Sales by Region for Each Publisher', ylabel='Total Sales', color=['#dac125', '#2b82d4', '#fe0192', '#9e8061'])
```

The table below shows the distribution of sales of video game publishers by region.

```python
print(sales.groupby('Publisher').sum().loc[:, ["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]])
```

The stacked bar chart shows that overall, Nintendo had the highest amount of total sales, in which sales was particularly the highest in North America (273.32 USD).

# Sales by Genre

Below shows a pie chart showing the percentage of sales each genre had in the North American Market.

Below shows a table of video games sales in North America by genre.

```python
colors = ['#d12e4e', '#e67019', '#f6ff00', '#1187ee', '#7429d6', '#c627d8', '#7f3e12', '#e61991']
top_sales = pd.read_csv("top100videogamesales.csv")

genre_grouping = top_sales.groupby('Genre').sum().sort_values(by='NA_Sales')
genre_regrouping = genre_grouping.drop(genre_grouping.index[[0,1,2]])

other = top_sales.groupby('Genre').sum().sort_values(by='NA_Sales').head(3)
other_genres = other.sum()
other_genres.name = "Other"

genre_regrouping.loc["Other"] = other_genres
genre_regrouping.plot(kind='pie', title='Sales of Video Games by Genre (North America)', y='NA_Sales', colors=colors, autopct='%1.1f%%', figsize=(7,7),  ylabel='')
```

The smaller slices (Puzzle, Fighting and Simulation) were grouped in the "Other" category.

```python
sales.groupby('Genre').sum().iloc[0:10, 5]
```

Below shows a pie chart of the percentage of sales each genre had in the European market.

```python
colors_eu = ['#e67019', '#d12e4e', '#f6ff00', '#1187ee', '#7429d6', '#7f3e12', '#c627d8', '#e61991']
top_sales_eu = pd.read_csv("top100videogamesales.csv")

genre_grouping_eu = top_sales_eu.groupby('Genre').sum().sort_values(by='EU_Sales')
genre_regrouping_eu = genre_grouping_eu.drop(genre_grouping_eu.index[[0,1,2]])

other_eu = top_sales.groupby('Genre').sum().sort_values(by='EU_Sales').head(3)
other_genres_eu = other.sum()
other_genres_eu.name = "Other"

genre_regrouping_eu.loc["Other"] = other_genres_eu
genre_regrouping_eu.plot(kind='pie', title='Sales of Video Games by Genre (Europe)', y='EU_Sales', colors=colors_eu, autopct='%1.1f%%', figsize=(7,7),  ylabel='')
```

Alike the pie chart for the video game sales by genre in North America, the smaller slices, being Puzzle, Fighting and Simulation were grouped together in the "Other Category".

Below shows a table of video game sales by genre in Europe.

```python
sales.groupby('Genre').sum().iloc[0:10, 6]
```

The distribution of sales of video games by genre in North America and Europe overall are very similar. For both continents, the genres that were grouped together in the "Other" category, the Puzzle, Fighting and Simulation games were the least popular. However, for the other genres, the distribution of sales was relatively even, though the top 2 genres for both continents were Sports and Shooter Games, in which Shooter games were the most popular in North America (24.8%), while Sports games were the most popular in Europe (19.7%).
